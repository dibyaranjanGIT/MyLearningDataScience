# Regular Expression for Pattern search in Python

```python
text = 'The agents phone number is 400-487-102'
```

```python
'phone' in text
```

```python
import re
```

```python
re.search('phone',text)
```

```python
pattern = 'My phone in first and phone in second'
```

```python
re.findall('phone',pattern)
```

```python
text = 'My phone number is 408-123-4485'
```

```python
re.search(r'\d\d\d-\d\d\d-\d\d\d\d',text)
```

```python
pattern = re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
match = re.search(pattern,text)
```

```python
#match.group(1)
```

```python
re.findall(r'\d{3}-\d{3}-\d{4}',text)
```

```python
re.search(r'cat|dog','the dog is here')
```

```python
re.search(r'.at','the cat with a hat sat there')
```

```python
re.findall(r'.at','the cat with a hat sat there')
```

```python
re.findall(r'2$','The sentence ends with 2')
```

```python
pattern = r'[\w]+-[\w]+'
```

```python
text = 'There are 3 dig-its in this te-xt 35 and 44 them is not 20'
```

```python
li = re.findall(pattern,text)
```
